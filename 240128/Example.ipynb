```python
import torch
import matplotlib.pyplot as plt

# Define the values of s1
s1_values = [0.2, 0.4, 0.6, 0.8]

# Define the range for c11 and c12
c11_c12_range = torch.linspace(0, 1, 100)

# Function to calculate c12 from c11 and s1 based on the equations
def calculate_c12(c11, s1, equation):
    if equation == 1:
        # Equation 1: c11 * s1 + c12 * (1 - s1) = s1
        return (s1 - c11 * s1) / (1 - s1)
    else:
        # Equation 2: c11 * s1 + c12 * (1 - s1) = 0.5
        return (0.5 - c11 * s1) / (1 - s1)

# Create plots with square boundaries and adjusted colors and styles
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i, s1 in enumerate(s1_values):
    ax = axs[i//2, i%2]

    # Calculate c12 values for each equation
    c12_values_eq1 = torch.tensor([calculate_c12(c11, s1, 1) for c11 in c11_c12_range])
    c12_values_eq2 = torch.tensor([calculate_c12(c11, s1, 2) for c11 in c11_c12_range])

    # Filter out values outside the 0 to 1 range
    valid_indices_eq1 = (c12_values_eq1 >= 0) & (c12_values_eq1 <= 1)
    valid_indices_eq2 = (c12_values_eq2 >= 0) & (c12_values_eq2 <= 1)

    ax.plot(c11_c12_range[valid_indices_eq1], c12_values_eq1[valid_indices_eq1], color='yellow', label='Equation 1')
    ax.plot(c11_c12_range[valid_indices_eq2], c12_values_eq2[valid_indices_eq2], color='red', label='Equation 2')

    # Draw a square boundary
    ax.plot([0, 1, 1, 0, 0], [0, 0, 1, 1, 0], color='black', linestyle='--')

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_title(f's1 = {s1}')
    ax.set_xlabel('c11')
    ax.set_ylabel('c12')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
c11 = 2
c12 = 1

# 
def equation1(s1):
    return c11 * s1 + c12 * (1 - s1)

def equation2(s1):
    return c11 * s1 + c12 * (1 - s1) - 0.5

#  s1 
s1_values = np.linspace(0, 1, 100)

# 12
equation1_values = equation1(s1_values)
equation2_values = equation2(s1_values)

# 
plt.plot(s1_values, equation1_values, label='Equation 1')
plt.plot(s1_values, equation2_values, label='Equation 2')
plt.axhline(0, color='black', linestyle='--', linewidth=1)  #  y=0 

#  s1 
s1_points = [0.2, 0.4, 0.6, 0.8]
for s1_point in s1_points:
    plt.scatter(s1_point, equation1(s1_point), color='red', marker='o')
    plt.scatter(s1_point, equation2(s1_point), color='blue', marker='o')

# 
plt.legend()
plt.xlabel('s1')
plt.ylabel('Function Value')
plt.title('Equations Visualization')
plt.grid(True)
plt.show()
```
