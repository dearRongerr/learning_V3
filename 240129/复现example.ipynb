```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_c12_equation1(c11, s1):
    return s1 - c11 * s1 / (1 - s1)

def calculate_c12_equation2(c11, s1):
    return (0.5 - c11 * s1) / (1 - s1)

# 
c11_values = np.linspace(0, 1, 100)
s1_values = [0.2, 0.4, 0.6, 0.8]

# 
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Equations 1 and 2 Visualization for Different s1 Values')

for i, s1 in enumerate(s1_values):
    row = i // 2
    col = i % 2

    c12_equation1_values = calculate_c12_equation1(c11_values, s1)
    c12_equation2_values = calculate_c12_equation2(c11_values, s1)

    axs[row, col].plot(c11_values, c12_equation1_values, label=f'Equation 1, s1={s1}')
    axs[row, col].plot(c11_values, c12_equation2_values, label=f'Equation 2, s1={s1}')
    axs[row, col].legend()
    axs[row, col].set_xlabel('$c_{11}$')
    axs[row, col].set_ylabel('$c_{12}$')
    axs[row, col].grid(True)

plt.show()
```
